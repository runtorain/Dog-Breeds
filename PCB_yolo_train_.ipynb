## Seperate datasets

```python
import os
import random
import shutil

# Define paths
dataset_dir = '/content/train'
train_dir = '/content/datasets/train'
val_dir = '/content/datasets/val'

# Create train and val directories if they don't exist
os.makedirs(os.path.join(train_dir, 'images'), exist_ok=True)
os.makedirs(os.path.join(train_dir, 'labels'), exist_ok=True)
os.makedirs(os.path.join(val_dir, 'images'), exist_ok=True)
os.makedirs(os.path.join(val_dir, 'labels'), exist_ok=True)

# Get list of all images
image_files = [f for f in os.listdir(os.path.join(dataset_dir, 'images')) if f.endswith('.jpg')]

# Set random seed for reproducibility
random.seed(42)

# Shuffle the list of images
random.shuffle(image_files)

# Define split ratio (adjust as needed)
split_ratio = 0.8  # 80% for training, 20% for validation

# Calculate split indices
split_index = int(len(image_files) * split_ratio)

# Split into train and val sets
train_images = image_files[:split_index]
val_images = image_files[split_index:]

# Move images from original directory to train and val directories
for image in train_images:
    img_name = image
    img_path_src = os.path.join(dataset_dir, 'images', img_name)
    img_path_dst = os.path.join(train_dir, 'images', img_name)
    shutil.copy(img_path_src, img_path_dst)

    label_name = img_name.replace('.jpg', '.txt')
    label_path_src = os.path.join(dataset_dir, 'labels', label_name)
    label_path_dst = os.path.join(train_dir, 'labels', label_name)
    shutil.copy(label_path_src, label_path_dst)

for image in val_images:
    img_name = image
    img_path_src = os.path.join(dataset_dir, 'images', img_name)
    img_path_dst = os.path.join(val_dir, 'images', img_name)
    shutil.copy(img_path_src, img_path_dst)

    label_name = img_name.replace('.jpg', '.txt')
    label_path_src = os.path.join(dataset_dir, 'labels', label_name)
    label_path_dst = os.path.join(val_dir, 'labels', label_name)
    shutil.copy(label_path_src, label_path_dst)

print("Dataset separated into train and val sets successfully.")
```

```python
import yaml

# Define the data dictionary
data = {
    'train': '/content/datasets/train/images',  # Path to training images directory
    'val': '/content/datasets/val/images',      # Path to validation images directory
    'nc': 6,  # Number of classes
    'names': ['missing_hole', 'mouse_bite', 'open_circuit', 'short', 'spur', 'spurious_copper']  # Class names
}

# Write the data dictionary to data.yaml
with open('/content/datasets/data.yaml', 'w') as yamlfile:
    yaml.dump(data, yamlfile, default_flow_style=False)

print("data.yaml created successfully.")
```

## Train Model

```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolov8m.pt")  # load a pretrained model (recommended for training)
```

```python
# Train the model
results = model.train(data="/content/datasets/data.yaml", epochs=20, imgsz=640)
```
